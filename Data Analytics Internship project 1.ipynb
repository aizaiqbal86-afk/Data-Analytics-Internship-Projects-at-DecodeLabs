# 13. Check Outliers

### Quantity

```python
Q1 = df["Quantity"].quantile(0.25)
Q3 = df["Quantity"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[(df["Quantity"] < lower) | (df["Quantity"] > upper)]

print(outliers.shape)
```
